```python
import numpy as np
import pandas as pd
import re
from collections import defaultdict
import string
import nltk

import gensim
from gensim.utils import simple_preprocess
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
data = pd.read_csv('train.csv')
pattern = r"<([^<>]*)>"
def to_lst(tag):
    return re.findall(pattern, tag)
data['Tags'] = data["Tags"].apply(to_lst)
data = data.drop(columns=['Id', 'CreationDate', 'Y'])
data
post_words = data.copy()
```

```python
post_words['code'] = post_words['Body'].map(lambda x: re.findall(r'(?s)(?<=<code>)(.*?)(?=<\/code>)', x))
post_words['Body_processed'] = post_words['Body'].map(lambda x: re.sub(r'(?s)(?<=<code>)(.*?)(?=<\/code>)','', x))
post_words['Body_processed'] = post_words['Body_processed'].map(lambda x: re.sub('<.*?>', '', x))
post_words['Body_processed'] = post_words['Body_processed'].map(lambda x: re.sub('\\n+', ' ', x))
post_words['Body_processed'] = post_words['Body_processed'].map(lambda x: x.lower())

post_words
```
